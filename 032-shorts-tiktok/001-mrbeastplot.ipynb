```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
df = pd.read_csv('MrBeast_youtube_stats.csv')
```

```python
df.groupby('title')['viewCount'].max() \
    .sort_values(ascending=False) \
    .head(10) \
    .sort_values() \
    .plot(kind='barh',
          figsize=(5, 10),
         title='Top MrBeast Videos')
plt.show()
```
