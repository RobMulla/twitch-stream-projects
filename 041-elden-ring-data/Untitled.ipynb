# Explode out JSON columns

```python
import pandas as pd
```

```python
df = pd.read_csv('elden-ring-data/weapons.csv')
```

```python
df['attack'].values[0]
```

```python
import matplotlib.pyplot as plt
```

```python
df.groupby('category')['weight'].plot(kind='kde')
plt.show()
```

```python
for i, d in df.iterrows():
    for c in d:
        try:
            if 'name' in c:
                try:
                    for x in eval(c):
                        if 'amount' in x.keys():
                            df.loc[i, x['name']] = x['amount']
                        elif 'scaling' in x.keys():
                            df.loc[i, x['name']] = x['scaling']
                        else:
                            print(x)
                except SyntaxError:
                    continue
        except TypeError:
            pass
```
