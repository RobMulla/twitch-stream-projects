# Matplotlib / Pandas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_parquet("flights/flights/Combined_Flights_2022.parquet")
fig, ax = plt.subplots(figsize=(5, 10))
df.groupby(["Airline"])["DepDelayMinutes"].mean().sort_values().plot(
    kind="barh", title="Average Flight Delay"
)
ax.set_xlabel("Average Delay Time")
```

# Seaborn

```python
import seaborn as sns
```

```python
fig, ax = plt.subplots(figsize=(5, 10))
sns.boxplot(data=df.query('0 < DepDelayMinutes < 60'),
            y='Airline', x='DepDelayMinutes')
ax.set_title('Flight Delay Time')
plt.show()
```

# Plotly Express

```python
import plotly.express as px
```

```python
fig = px.scatter(df.query('Operating_Airline in ("WN","DL","AA")') \
                     .sample(5000),
            x='ArrDelayMinutes',
            y='Distance',
           color='Airline')         
fig.update_layout(legend=dict(
    orientation="h",
    yanchor="bottom",
    y=1.02,
    xanchor="right",
    x=1
))
```
