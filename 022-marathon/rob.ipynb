```python
import pandas as pd
```

```python
df = pd.read_csv('Rob-OneCityMarathonTraining - results.csv')
```

```python
df['chip_minutes'] = (pd.to_datetime(df['CHIP TIME']) - pd.to_datetime(pd.to_datetime(df['CHIP TIME']).dt.date)).dt.total_seconds() / 60
```

```python
ax = df['chip_minutes'].plot(kind='hist', bins=50, figsize=(15, 10))
ax.axvline(df.loc[df['NAME'].str.contains('Mulla')]['chip_minutes'].values[0], color='black', ls='--')
```

```python
df2 = df.loc[df['M/F'] == "M"]
ax = df2['chip_minutes'].plot(kind='hist', bins=40, figsize=(15, 10))
ax.axvline(df2.loc[df2['NAME'].str.contains('Mulla')]['chip_minutes'].values[0], color='black', ls='--')
```
