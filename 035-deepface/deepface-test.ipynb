# Emotion from Webcam

```python
import pandas as pd
import cv2
import matplotlib.pyplot as plt
from deepface import DeepFace
```

```python
cap = cv2.VideoCapture(1)
res, img = cap.read()
fig, ax = plt.subplots(figsize=(10, 10))

ax.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
cap.release()
```

```python
analysis = DeepFace.analyze(img, actions=['emotion'])
```

```python
results = pd.DataFrame().from_dict(analysis['emotion'], orient='index', columns=['score'])
```

```python
results.sort_values('score').plot(kind='barh', figsize=(5, 8), title='Emotion')
```
