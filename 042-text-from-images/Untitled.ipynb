# Format JSON

```python
import pandas as pd
import json
```

```python
%load_ext lab_black
```

```python
df = pd.read_json('TextOCR_0.1_train.json')
```

```python
annot = "TextOCR_0.1_train.json"


with open(annot) as data_file:
    data = json.load(data_file)
```

```python
df = pd.DataFrame(data["anns"])
```

```python
df_annot = df.T
```

```python
df_annot = df_annot.reset_index(drop=True)
```

```python
df_annot.to_csv("annot.csv")
df_annot.to_parquet("annot.parquet")
```

```python
data.keys()
```

```python
df_imgs = pd.DataFrame(data["imgs"]).T
df_imgs = df_imgs.reset_index(drop=True)
```

```python
df_imgs.to_csv("img.csv")
df_imgs.to_parquet("img.parquet")
```
